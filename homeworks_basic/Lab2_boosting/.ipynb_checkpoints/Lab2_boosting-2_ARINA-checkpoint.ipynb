```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import numpy as np
```

```python
X_train = np.genfromtxt('data/train/X_train.txt')
y_train = np.genfromtxt('data/train/y_train.txt')

X_test = np.genfromtxt('data/test/X_test.txt')
y_test = np.genfromtxt('data/test/y_test.txt')

with open('data/activity_labels.txt', 'r') as iofile:
    activity_labels = iofile.readlines()

activity_labels = [x.replace('\n', '').split(' ') for x in activity_labels]
activity_labels = dict([(int(x[0]), x[1]) for x in activity_labels])
```

```python
activity_labels
```

```python
print(X_train.shape)
data_mean = X_train.mean(axis=0)
data_std = X_train.std(axis=0)

X_train = (X_train - data_mean)/data_std
X_test = (X_test - data_mean)/data_std
```

The dataset has some duplicating features. File `unique_columns.txt` stores the indices of the unique ones. 

```python
try: 
    unique_columns = np.genfromtxt('unique_columns.txt', delimiter=',').astype(int)
except FileNotFoundError:
    ! wget https://raw.githubusercontent.com/ml-mipt/ml-mipt/basic_s20/homeworks_basic/Lab2_boosting/unique_columns.txt -nc
    unique_columns = np.genfromtxt('unique_columns.txt', delimiter=',').astype(int)

X_train_unique = X_train[:, unique_columns]
X_test_unique = X_test[:, unique_columns]
```

PCA could be useful in this case. E.g.

```python
X_train_pca = pca.fit_transform(X_train_unique)
X_test_pca = pca.transform(X_test_unique)
```

### Models

```python
xgboost_model = xgboost.XGBClassifier()
param_dist = {'max_depth': np.arange(3, 15), 
              'n_estimators': np.arange(10, 300, 30),
              'learning_rate': np.linspace(0.05, 0.8, 20)}
rand_search = RandomizedSearchCV(estimator=xgboost_model, scoring='accuracy', cv=5, 
                                 param_distributions=param_dist, n_iter=20)
rand_search.fit(X_train_pca, y_train)
```

_We got resulting models sorted from best down to worst using our parameters_

```python
results = pd.DataFrame(rand_search.cv_results_)
results.sort_values(by='rank_test_score', inplace=True)
results.head()
```

_Lets take 3 best results_

```python
head = results.head(3)

p1 = head.loc[:, 'params'].iloc[0]
rs1 = xgboost.XGBClassifier(n_estimators = p1['n_estimators'], 
                            max_depth = p1['max_depth'], learning_rate = p1['learning_rate'])

rs1.fit(X_train_pca, y_train)
print(accuracy_score(rs1.predict(X_test_pca), y_test))
```

```python
p2 = head.loc[:, 'params'].iloc[1]
rs2 = xgboost.XGBClassifier(n_estimators = p2['n_estimators'], 
                            max_depth = p2['max_depth'], learning_rate = p2['learning_rate'])

rs2.fit(X_train_pca, y_train)
print(accuracy_score(rs2.predict(X_test_pca), y_test))
```

```python
p3 = head.loc[:, 'params'].iloc[2]
rs3 = xgboost.XGBClassifier(n_estimators = p3['n_estimators'], 
                            max_depth = p3['max_depth'], learning_rate = p3['learning_rate'])

rs3.fit(X_train_pca, y_train)
print(accuracy_score(rs3.predict(X_test_pca), y_test))
```

_Moreover lets show the dependency of accuracy on the overlooked parameters_

```python
[p1, p2, p3]
```

```python
plot_roc(y_test, rs1.predict_proba(X_test_pca), figsize = (8, 6))
plot_roc(y_test, rs2.predict_proba(X_test_pca), figsize = (8, 6))
plot_roc(y_test, rs3.predict_proba(X_test_pca), figsize = (8, 6))
```

_... now show accuracy dependency on the plots_

```python
estims = [10, 50, 100, 150, 180, 200, 220, 250, 300, 500]
acc_estim = []
for i in estims:
    xgb = xgboost.XGBClassifier(n_estimators = i)
    xgb.fit(X_train_pca, y_train)
    acc_estim.append(accuracy_score(xgb.predict(X_test_pca), y_test))
```

```python
plt.figure(figsize=(6, 4))
plt.plot(estims, acc_estim, lw=5)
plt.xlabel('amount of estimators')
plt.ylabel('accuracy')
plt.title('Dependency between estimators and accuracy')
plt.show()
```

_As we can see: the bigger amount of estimators we take the better is accuracy, but also we can notice that starting from about 200~210 accuracy changes are small but it takes more resources then its better to use n_estimators parameter on about that value_

```python
depths = [1, 3, 5, 7, 10, 12, 15, 20, 25]
acc_dep = []
for i in depths:
    xgb = xgboost.XGBClassifier(max_depth = i)
    xgb.fit(X_train_pca, y_train)
    acc_dep.append(accuracy_score(xgb.predict(X_test_pca), y_test))
```

```python
plt.figure(figsize=(6, 4))
plt.plot(depths, acc_dep, lw=5)
plt.xlabel('depth')
plt.ylabel('accuracy')
plt.title('Dependency between maximal depth and accuracy')
plt.show()
```

_As we now see big depth is not good so we take best accuracy and less depth it is around 4_

```python
rates = np.linspace(0.2, 0.7, 10)
acc_rate = []
for i in rates:
    xgb = xgboost.XGBClassifier(learning_rate = i)
    xgb.fit(X_train_pca, y_train)
    acc_rate.append(accuracy_score(xgb.predict(X_test_pca), y_test))
```

```python
plt.figure(figsize=(6, 4))
plt.plot(rates, acc_rate, lw=5)
plt.xlabel('learning rate')
plt.ylabel('accuracy')
plt.title('Dependency between learning rate and accuracy')
plt.show()
```

_It seems that there is no dependency but still we see that near 0.47 is the best accuracy on given linspace with other parameters set default and we can suspect that this very parameter should be searched together with othes_

### Part 2. Blending the models

Take three (or more) best models and try to build the blending ensemble of them. Compare the quality of the final model using the same quality measures as above.

_Lets use library stacking as blending is just a special case of stacking_

```python
rs1 = xgboost.XGBClassifier(n_estimators = p1['n_estimators'], 
                            max_depth = p1['max_depth'], learning_rate = p1['learning_rate'])
rs2 = xgboost.XGBClassifier(n_estimators = p2['n_estimators'], 
                            max_depth = p2['max_depth'], learning_rate = p2['learning_rate'])
rs3 = xgboost.XGBClassifier(n_estimators = p3['n_estimators'], 
                            max_depth = p3['max_depth'], learning_rate = p3['learning_rate'])
```

```python
best_models = [rs3, rs2, rs1]
stacked_train, stacked_test = stacking( best_models, X_train, y_train, X_test, regression = False, 
                                       mode = 'oof_pred_bag', needs_proba = False, save_dir = None, 
                                       metric = accuracy_score, n_folds = 5, stratified = True, 
                                       shuffle = True, random_state = 0 )    
```

```python
model = xgboost.XGBClassifier()
model.fit(stacked_train, y_train)
acc = accuracy_score(model.predict(stacked_test), y_test)
acc
```

_As we see accuracy rised after using stackinging idea, because we can value better models with bigger weights and get something better than just mean value through ensemble_
